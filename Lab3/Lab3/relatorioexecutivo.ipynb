## Relatório Executivo de Clusterização de Clientes
#### Resumo executivo

  A clusterização de clientes do cartão de crédito revelou perfis distintos de comportamento relevantes para estratégia de negócios. Utilizando algoritmos K-Means, DBSCAN e Clusterização Hierárquica, encontramos grupos bem separados e interpretáveis, além de identificar outliers e oportunidades. O K-Means foi o método principal recomendado, pela boa separação dos grupos (Davies-Bouldin Index de 1.63, Calinski-Harabasz Score de 11.891) e fácil replicação. Cada perfil descoberto permite ações direcionadas: fidelização de pagadores, educação de parceladores, estímulo para inativos e mitigação de riscos em clientes fora do padrão.

1. Metodologia utilizada
- Pré-processamento: Correção de valores ausentes (mediana) e normalização MinMax.

- Modelagem:

  - K-Means: Determinação do número ideal de clusters por método do cotovelo e Silhouette Score.

  - DBSCAN: Teste de diferentes eps/min_samples, identificação de outliers reais ou potenciais erros nos dados.

  - Clusterização Hierárquica: Análise com vários métodos de linkage, decisão sobre corte no dendrograma.

- Validações:

  - Silhouette Score, Davies-Bouldin Index (menor é melhor), e Calinski-Harabasz (maior é melhor).

- Visualizações: Redução de dimensionalidade via PCA para visualização dos grupos em 2D e inspeção visual dos clusters.

2. Resultados e principais visualizações
- K-Means:

  - 5 clusters definidos, bem separados e de fácil interpretação, com qualidade comprovada por métricas e visualização PCA.

- DBSCAN:

  - Detectou ~5% outliers, permitindo flagrar clientes fora do padrão para investigação ou ações específicas.

  - Ajustes em eps e min_samples alteram sensibilidade, equilibrando número de clusters e outliers.

- Hierárquica:

  - Dendrogramas confirmam existência de 4-5 grandes grupos, alinhando-se ao resultado do K-Means.​

3. Descrição dos clusters (exemplo):
- Cluster 0: Compradores à Vista – Alto saldo, muitas compras à vista, pagamentos integrais frequentes.

- Cluster 1: Parceladores Ativos – Maior uso do parcelamento, gasto recorrente.

- Cluster 2: Sacadores – Saca dinheiro frequentemente, baixo uso para compras.

- Cluster 3: Pagadores Integrais – Menor uso do cartão, mas pagamento sempre em dia e completo.

- Cluster 4: Inativos/Irregulares – Baixo engajamento, possível desinteresse ou uso esporádico.

4. Recomendações de negócio
- Fidelize Pagadores Integrais: Ofereça cartões especiais, programas de recompensa e upgrades para quem mantém pagamento em dia.

- Acompanhe Parceladores: Implemente alertas para risco de crédito, oportunidades de renegociação, promoções de parcelamento sem juros.

- Eduque Sacadores: Dê opções de microcrédito para reduzir custos e risco de inadimplência pelo uso intenso de saque.

- Reative Inativos: Campanhas de incentivo, promoções direcionadas e abordagem personalizada para clientes quase perdidos.

- Analise Outliers: Investigue para identificar fraudes, oportunidades VIP, ou erros de cadastro/dados.

5. Considerações finais

  A clusterização permitiu criar segmentações personalizadas, facilitando o desenho de ofertas, estímulo ao uso, prevenção de riscos e aumento de receita por cliente. Recomendo repetir periodicamente as análises, atualizando os perfis e monitorando mudanças no comportamento dos clientes, fazendo da clusterização uma prática estratégica contínua.
